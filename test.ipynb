```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from tkinter import *

def predict_virus():
    # Kullanıcının girdiği bilgileri al
    protein1 = entry_protein1.get()
    protein2 = entry_protein2.get()
    host_organism = entry_host_organism.get()
    miscore = entry_miscore.get()

    # Kullanıcının girdiği bilgileri kullanarak tahmin yap
    input_data = f'{protein1} {protein2} {host_organism} {miscore}'
    input_vectorized = vectorizer.transform([input_data])
    prediction = model.predict(input_vectorized)

    # Sonucu ekrana yazdır
    label_result.config(text=f'Tahmin Edilen Virüs: {prediction[0]}')

# Verisetini yükle
data = pd.read_csv('processed_datasetSon.csv')

# Eksik değerleri doldur
data = data.fillna('')

# Bağımsız değişken (X) ve bağımlı değişken (y) olarak ayır
X = data['Protein1'] + ' ' + data['Protein2'] + ' ' + data['HostOrganism'] + ' ' + data['Miscore'].astype(str)
y = data['VirusName']

# Metin verilerini vektörlere dönüştür
vectorizer = CountVectorizer()
X_vectorized = vectorizer.fit_transform(X)

# Naive Bayes modelini oluştur ve eğit
model = MultinomialNB()
model.fit(X_vectorized, y)

# Tkinter arayüzü oluştur
root = Tk()
root.title("Makine Öğrenmesi Projesi")

# Arka plan rengini ayarla
root.configure(bg='#add8e6')

# Başlık fotoğrafı ekle
photo = PhotoImage(file='virus.png')
label_photo = Label(root, image=photo, bg='#add8e6')
label_photo.grid(row=0, column=2, rowspan=6, padx=10)

# Giriş widget'ları
label_protein1 = Label(root, text="Protein1:", bg='#add8e6')
label_protein1.grid(row=0, column=0, sticky=E)
entry_protein1 = Entry(root)
entry_protein1.grid(row=0, column=1)

label_protein2 = Label(root, text="Protein2:", bg='#add8e6')
label_protein2.grid(row=1, column=0, sticky=E)
entry_protein2 = Entry(root)
entry_protein2.grid(row=1, column=1)

label_host_organism = Label(root, text="Host Organism:", bg='#add8e6')
label_host_organism.grid(row=2, column=0, sticky=E)
entry_host_organism = Entry(root)
entry_host_organism.grid(row=2, column=1)

label_miscore = Label(root, text="Miscore:", bg='#add8e6')
label_miscore.grid(row=3, column=0, sticky=E)
entry_miscore = Entry(root)
entry_miscore.grid(row=3, column=1)

# Tahmin butonu
button_predict = Button(root, text="Predict Virus", command=predict_virus, bg='#87CEEB')
button_predict.grid(row=4, column=0, columnspan=2, pady=10)

# Sonuç label'ı
label_result = Label(root, text="", bg='#add8e6')
label_result.grid(row=5, column=0, columnspan=2)

root.mainloop()
```
